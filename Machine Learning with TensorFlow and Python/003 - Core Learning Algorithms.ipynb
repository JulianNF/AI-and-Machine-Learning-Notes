### Linear Regression

One of the most basic forms of machine learning, used to predict numeric values.

A best-fit line is derived for the dataset, and the function that defines this line can then be used to predict unknown or future value.

For a linear regression, we have a function $f(x) = mx + b$

Note that linear regression can also be used with a 3-axis dataset / graph, so long as the dataset values do correlate linearly.

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 2.5, 3, 4]
y = [1, 4, 7, 9, 15]

plt.title('An example of a linear regression best-fit line in matplotlib')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x, y, 'ro')  # plot the data points
plt.axis([0, 6, 0, 20])

# 1 for linear, 2 for quadratic, 3 for cubic, ...
polynomial = np.polyfit(x, y, 1)
slope = polynomial[0]
y_intercept = polynomial[1]

plt.plot(  # plot the best-fit line
    np.unique(x),
    slope*np.unique(x) + y_intercept
    # np.poly1d(slope_and_yIntersect)(np.unique(x))
)

plt.show()
```
